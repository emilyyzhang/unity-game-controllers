```python
##USAGE: should only be executed by Makefile, e.g. "make cov_matrix"


import os
import numpy as np

import matplotlib.pyplot as plt
import scipy.stats

myCurriculum = [p for p in dir(Curriculum)
                   if isinstance(getattr(Curriculum, p), str)
                   and not p.startswith('__')]

loaded_weighted_lev_distances = np.load(os.getcwd() + '/../GameUtils/PronunciationUtils/data/weighted_lev_distances.npy')
loaded_gloVe_distances = np.load(os.getcwd() + '/../GameUtils/PronunciationUtils/data/glove_distances.npy')

### THESE MUST ADD UP TO 1!!! ###
PHONETIC_WEIGHT = .33
SEMANTIC_WEIGHT = .66

lev_ds = []
glove_ds = []
covariance_matrix = []
def create_cov_matrix():    

    for i in range(0, len(myCurriculum)):
        for j in range(0, len(myCurriculum)):

            #yeah I know we're calculating it twice - TODO: Optimize
            lev_ds.append(1 - loaded_weighted_lev_distances[i][j])
            glove_ds.append(loaded_gloVe_distances[i][j])
            covariance_matrix.append((PHONETIC_WEIGHT * (1 - loaded_weighted_lev_distances[i][j])) + \
                                        (SEMANTIC_WEIGHT * loaded_gloVe_distances[i][j]))

    fig, plts = plt.subplots(1, 3, figsize=(15, 10))

    print("MEAN VAR MED of Lev")
    print(np.mean(lev_ds))
    print(np.std(lev_ds))
    print(np.median(lev_ds))
    plts[0].scatter(lev_ds, range(len(lev_ds)))

    print("MEAN VAR MED of Glove")
    print(np.mean(glove_ds))
    print(np.std(glove_ds))
    print(np.median(glove_ds))
    plts[1].scatter(glove_ds, range(len(glove_ds)))
    
    
    print("MEAN VAR MED of COV")
    print(np.mean(covariance_matrix))
    print(np.std(covariance_matrix))
    print(np.median(covariance_matrix))
    plts[2].scatter(covariance_matrix, range(len(covariance_matrix)))
    plt.draw()
    plt.show(block=False)
      

create_cov_matrix()
```
